##### Probamos ya nuestro codigo en todo el dataset

```python
def calculate_rsi(data, tickers, list_window):

    df_temp = []

    for ticker in tickers:

        df_ticker = data_copy[data_copy['Ticker'] == ticker]
        

        for num in list_window:

            delta = df_ticker.Close.diff(num)

            gain = (delta.where(delta > 0, 0)).rolling(window=num).mean()
            
            loss = (-delta.where(delta < 0, 0)).rolling(window=num).mean()
            
            rs = gain / loss
            rsi = 100 - (100 / (1 + rs))
            
            df_ticker[f'RSI_{num}'] = rsi

            #print(ticker,df_ticker)
        df_temp.append(df_ticker) 

    #print('Este es el df temporal:',df_temp)
    return df_temp
```

#### Cálculo del EWMA (Exponentially Weighted Moving Average -- >Media Movil Exponencial Ponderada)

1. Este indicador se utiliza para calcular e interpretar la Volatilidad Histórica. 

2. El EWMA es como calcular la desviación estándar de los rendimientos del activo, pero añadiendo algunos factores que hacen que sea un indicador más ajustado y preciso para calcular esa volatilidad. Estos añadidos son:

    - Ponderación exponencial: conla formula EWMA se ponderan a la alza los datos más recientes
    
    - En la desviación estandar, simplemente se toman los datos del dia anterior. Con EWMA podemos calcular esa volatilidad en periodos de 7, 50 y 200

3. El primer valor de cada ewma debería ser el mismo que Close para cada ETF

##### Código funcion para dataset completo

```python
def calculate_ewma(data,tickers,list_span):
        df_temp = []

        for ticker in tickers:
                df_ticker = data[data.Ticker == ticker] 
                #print(df_ticker)
                for num in list_span:
                        #print(df_ticker['Close'].ewm(span=7, adjust=False).mean())
                        df_ticker[f'ewma_{num}'] = df_ticker['Close'].ewm(span=num, adjust=False).mean()

                #print('ESTE ES DF_TICKER',df_ticker)

                df_temp.append(df_ticker)

        return df_temp
```

```python
data = data_copy
list_span = [7,50,200]
tickers = data_copy.Ticker.unique()

etfs_2018_2024_ewmas = calculate_ewma(data,tickers,list_span)
```

```python
etfs_2018_2024_ewmas = pd.concat(etfs_2018_2024_ewmas)
```

```python
from functions import calculate_ewma

def calculate_ewma(data,tickers,list_span):
        df_temp = []

        for ticker in tickers:
                df_ticker = data[data.Ticker == ticker] 
                #print(df_ticker)
                for num in list_span:
                        #print(df_ticker['Close'].ewm(span=7, adjust=False).mean())
                        df_ticker[f'ewma_{num}'] = df_ticker['Close'].ewm(span=num, adjust=False).mean()

                #print('ESTE ES DF_TICKER',df_ticker)

                df_temp.append(df_ticker)

        return df_temp

def calculate_macd(data,list_span):
    
    columns_to_drop = data.columns

    tickers = ['FGDL','SSO']
    emas_12_26 = calculate_ewma(data,tickers,list_span) #devuelve una lista con las columnas
    emas_12_26 = pd.concat(emas_12_26)
    emas_12_26 = emas_12_26.drop(columns_to_drop,axis=1)
    #print('mis emas:',emas_12_26['ewma_12'])
    
    macd = emas_12_26['ewma_12'] - emas_12_26['ewma_26']
    #print('MACD',macd)

    signal = macd.ewm(span=9, adjust=False).mean()
    #print('SIGNAL', signal)
    data['MACD'] = signal
    

    return data

```
