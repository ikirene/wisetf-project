```python
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
from sklearn.model_selection import train_test_split
import pycaret
```

```python
data = pd.read_csv('etfs_dataset_modelling.csv')
```

```python
data.dropna(inplace=True)
```

```python
data = data.drop(['Unnamed: 0', 'Ticker', 'Date'], axis=1)
data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['Open','High','Low','Close','Volume', 'Dividends','Stock Splits', 'RSI_9','RSI_14','RSI_25','ewma_7','ewma_50','ewma_200']] = scaler.fit_transform(data[['Open','High','Low','Close','Volume', 'Dividends','Stock Splits', 'RSI_9','RSI_14','RSI_25','ewma_7','ewma_50','ewma_200']])
```

```python
from pycaret.regression import *

model1 = setup(data = data, target = data['Close_target'],train_size=0.7,preprocess=False, transformation=False,
               remove_multicollinearity=True, multicollinearity_threshold=0.9, feature_selection=True)

best = compare_models()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np

X = data.drop(['Close_target'], axis=1)
y = data['Close_target']

X_train,X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)

model2 = LinearRegression()

model2.fit(X_train, y_train)
y_pred = model2.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

results_model2 = {'Model': ['Linear Regression'],'RMSE': [rmse],'R2': [r2] }

results_df = pd.DataFrame(results_model2)
results_df
```

```python
X = data.drop(['Close_target','RSI_9','RSI_14','RSI_25','ewma_7','ewma_50','ewma_200'], axis=1)
y = data['Close_target']

X_train,X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)

model2 = LinearRegression()

model2.fit(X_train, y_train)
y_pred = model2.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

results_model2 = {'Model': ['Linear Regression'],'RMSE': [rmse],'R2': [r2] }

results_df = pd.DataFrame(results_model2)
results_df
```
